# **Pipeline de Dados do Telegram pela AWS**

O projeto tem como proposta demonstrar como implantar um **pipeline** com os dados de **texto** gerados em um grupo do aplicativo de mensagens **Telegram**. Os dados serão processados em diversas etapas, melhorando sua utilidade para que sejam preparados para análise através dos serviços da **Amazon Web Services**.

A **arquitetura** do projeto é dividida em **transacional**, onde os dados são produzidos (no Telegram) e **analítica** (na Amazon Web Services - AWS) onde ocorrerá todo o processamento, manipulação, armazenamento e análise dos dados.

No grupo do Telegram (etapa transacional) as mensagens enviadas pelos usuários serão capturadas através da API (Application Programming Interface) de bots dos Telegram e redirecionadas através de um Webhook para a API Gateway da AWS.

Na AWS (etapa analítica) essas mensagens passarão pelas fases de ingestão, onde os dados serão armazenados em seu estado bruto; pela fase de ETL (extract, transform e load / extração, transformação e carregamento) onde os dados passarão por uma limpeza, um refinamento e armazenados; e pela fase de apresentação, onde será criada uma tabela em SQL com os dados já refinados e prontos para serem analisados.

# <span style="background-color: #fff001 ; color:#000; padding: 10px ; display: block;font-size:0.8em; font-weight: bold;">  INGESTÃO </span>

Esta etapa é responsável, pela ingestão dos dados transacionais em ambientes analíticos. As mensagens coletadas pelo bots através da API web de bots do Telegram serão ingeridas via streaming (conforme são produzidas) através de um webhook (gancho web) que irá redirecionar as mensagens para outra API web.

Para isso precisaremos de um serviço que forneça um API web que irá receber os dados redirecionados, o AWS API Gateway. Por sua vez, esses dados recebidos serão redirecionados para outro serviço da AWS, o AWS Lambda que irá armazenar o dado em seu formato original (JSON) em um bucket do AWS S3.

Vamos trabalhar de forma inversa a que os dados serão processados, pois para redirecionar precisaremos de dados das etapas seguintes, então faremos o seguinte:

* Criaremos um bucket no AWS S3;
* Criaremos uma função no AWS Lambda;
* Criaremos uma API web no AWS API Gateway;
* Configuraremos o webhook da API de bots do Telegram.

### **AWS S3**

O AWS S3 armazenará as mensagens captadas pelo bot do Telegram em um bucket que seguindo o padrão de data lake (repositório com grande volume dados), onde a organização dos dados armazenados é dada de acordo com os níveis de processamento, o bucket terá o sufixo -raw em seu nome pelo fato de os dados estarem em seu formato original: JSON

### **AWS Lambda**

No AWS Lambda escreveremos uma função que receberá as mensagens do Telegram via AWS API Gateway através do parametro event, verificará se a mensagem veio do grupo correto, se sim a escreverá no bucket do AWS S3 em seu formato original JSON e retornará uma mensagem de sucesso.

````python
import os
import json
import logging
from datetime import datetime, timedelta, timezone

import boto3

def lambda_handler(event: dict, context: dict) -> dict:

  # vars de ambiente

  BUCKET = os.environ['AWS_S3_BUCKET']
  TELEGRAM_CHAT_ID = int(os.environ['TELEGRAM_CHAT_ID'])

  # vars lógicas

  tzinfo = timezone(offset=timedelta(hours=-3))
  date = datetime.now(tzinfo).strftime('%Y-%m-%d')
  timestamp = datetime.now(tzinfo).strftime('%Y%m%d%H%M%S%f')

  filename = f'{timestamp}.json'

  # código principal

  client = boto3.client('s3')

  try:
    message = json.loads(event["body"])
    chat_id = message["message"]["chat"]["id"]

    if chat_id == TELEGRAM_CHAT_ID:

      with open(f"/tmp/{filename}", mode='w', encoding='utf8') as fp:
        json.dump(message, fp)

      client.upload_file(f'/tmp/{filename}', BUCKET, f'telegram/context_date={date}/{filename}')

  except Exception as exc:
      logging.error(msg=exc)
      return dict(statusCode="500")

  else:
      return dict(statusCode="200")
````

**- Variáveis de ambiente**

Observe que o código exige a configuração de duas variáveis de ambiente:
* AWS_S3_BUCKET que terá o nome do bucket criado no AWS S3
* TELEGRAM_CHAT_ID que terá o id do chat do grupo do Telegram (que obtivemos através do método getUpdades).

Para adicionar variáveis de ambiente, no console da função da função Lambda, basta acessar:

# <span style="background-color: #f8b243 ; color:#000; padding: 10px ; display: block;font-size:0.8em; font-weight: bold;">  ETL </span>


Esta é a etapa de extração, transformação e carregamento (extraction, transformation and load) e é responsável pelo refinamento dos dados.

No Projeto, nesta etapa criaremos uma regra no AWS EVENTBRIDGE que acionará diariamente uma função no AWS Lambda, função essa que irá manipular os dados pois passarão por um processo de data wrangling onde serão filtrados e salvará o conteúdo processado em um arquivo particionado no formato Apache Parquet em um bucket no AWS S3

Para isso, vamos:

* Criar um bucket com os dados enriquecidos processados
* Criar uma função do AWS Lambda como motor de processamento
* Configurar uma regra do AWS Event Bridge como gatilho diáro da função

### **AWS S3**

No AWS S3 serão armazenadas as mensagens diárias processadas em um único arquivo no formato Parquet.

Para a criação de um bucket, como padrão, vamos adicionar o sufixo -enriched ao seu nome



### **AWS Lambda**

Na etapa de ETL, a função Lambda carregará as mensagens JSON do dia anterior (D-1) armazenadas no bucket de dados crus, executará uma função de data wrangling, criará uma tabela do pyarrow e concatenará com as demais e salvará o resultado em um único arquivo PARQUET em um bucket enriquecido e também particionado por dia.

Não esquecendo de configurar as variáveis de ambiente, de adicionar as permissões de interação com AWS S3 no AWS IAM, de configurar o timeout e de adicionar o layer (camada) com o código do pacote Python PyArrow.

* Definir as variáveis de ambiente em configurações:

  * AWS_S3_BUCKET - nome do bucket cru (com fim -raw)
  * AWS_S3_ENRICHED - nome do bucket enqriquecido (sufixo -enriched)


* Adicionar as Permissões em Configurações no console da função

* Aumentar o tempo de processamento em Configurações < Configuraçãos Geral

* Criar uma camada para injetar o Pyarrow na função Lambda:

  * No AWS S3 criar um bucket
  * Fazer download do pacote Python do PyArrow [neste link](https://github.com/awslabs/aws-data-wrangler/releases) e fazer o upload para o bucket;

 * Criar uma camada personalizada e conectar na função.

### Código da função inserido no AWS Lambda

````python
import os
import json
import logging
from datetime import datetime, timedelta, timezone

import boto3
import pyarrow as pa
import pyarrow.parquet as pq

def lambda_handler(event: dict, context: dict) -> bool:

  # vars de ambiente

  RAW_BUCKET = os.environ['AWS_S3_BUCKET']
  ENRICHED_BUCKET = os.environ['AWS_S3_ENRICHED']

  # vars lógicas

  tzinfo = timezone(offset=timedelta(hours=-3))
  # date = (datetime.now(tzinfo) - timedelta(days=0)).strftime('%Y-%m-%d')
  date = (datetime.now(tzinfo) - timedelta(days=1)).strftime('%Y-%m-%d')
  timestamp = datetime.now(tzinfo).strftime('%Y%m%d%H%M%S%f')

  # código principal

  table = None
  client = boto3.client('s3')

  try:

      response = client.list_objects_v2(Bucket=RAW_BUCKET, Prefix=f'telegram/context_date={date}')

      for content in response['Contents']:

        key = content['Key']
        client.download_file(RAW_BUCKET, key, f"/tmp/{key.split('/')[-1]}")

        with open(f"/tmp/{key.split('/')[-1]}", mode='r', encoding='utf8') as fp:

          data = json.load(fp)
          data = data["message"]

        parsed_data = parse_data(data=data)
        iter_table = pa.Table.from_pydict(mapping=parsed_data)

        if table:

          table = pa.concat_tables([table, iter_table])

        else:

          table = iter_table
          iter_table = None

      pq.write_table(table=table, where=f'/tmp/{timestamp}.parquet')
      client.upload_file(f"/tmp/{timestamp}.parquet", ENRICHED_BUCKET, f"telegram/context_date={date}/{timestamp}.parquet")

      return True

  except Exception as exc:
      logging.error(msg=exc)
      return False

# Função Data Wrangling

def parse_data(data: dict) -> dict:

  date = datetime.now().strftime('%Y-%m-%d')
  timestamp = datetime.now().strftime('%Y-%m-%d %H:%M:%S')

  parsed_data = dict()

  for key, value in data.items():

      if key == 'from':
          for k, v in data[key].items():
              if k in ['id', 'is_bot', 'first_name']:
                parsed_data[f"{key if key == 'chat' else 'user'}_{k}"] = [v]

      elif key == 'chat':
          for k, v in data[key].items():
              if k in ['id', 'type']:
                parsed_data[f"{key if key == 'chat' else 'user'}_{k}"] = [v]

      elif key in ['message_id', 'date', 'text']:
          parsed_data[key] = [value]

  if not 'text' in parsed_data.keys():
    parsed_data['text'] = [None]

  return parsed_data
````

Obs.: Para testar a função, substitua na variável lógica o código `timedelta(days=1)` por `timedelta(days=0)`, permitindo assim o processamento de mensagens do mesmo dia.

## **Análise dos Dados**

Com o dado disponível, o usuário pode executar as mais variadas consultas analíticas. Seguem alguns exemplos:

**Quantidade de mensagens por dia, agrupadas por data:**

```
SELECT
  context_date,
  count(1) AS "message_amount"
FROM "telegram"
GROUP BY context_date
ORDER BY context_date DESC
```

**Quantas vezes uma determinada palavra aparece na mensagens, no caso: *testando*, na coluna *text*:**

```
SELECT COUNT(*) AS "contagem",
text,
context_date
FROM "telegram"
WHERE "text" LIKE '%testando%'
GROUP BY context_date,
text
```

**Média do comprimento das mensagens diárias dos usuários:**

```
SELECT
  user_first_name,
  context_date,
  CAST(AVG(length(text)) AS INT) AS "media_comp_mesg"
FROM "telegram"
GROUP BY
  user_first_name,
  context_date
ORDER BY context_date DESC
```

**Quantidade de mensagens por hora por dia da semana por número da semana ordenados por semana do ano:**

Onde:
* hora = de 0 a 23
* semana = de 1 (seg) a 7 (dom)
* sem do ano = de 1 a 52

```
WITH
parsed_date_cte AS (
    SELECT
        *,
        CAST(date_format(from_unixtime("date"),'%Y-%m-%d %H:%i:%s') AS timestamp) AS parsed_date
    FROM "telegram"
),
hour_week_cte AS (
    SELECT
        *,
        EXTRACT(hour FROM parsed_date) AS parsed_date_hour,
        EXTRACT(dow FROM parsed_date) AS parsed_date_weekday,
        EXTRACT(week FROM parsed_date) AS parsed_date_weeknum
    FROM parsed_date_cte
)
SELECT
    parsed_date_hour,
    parsed_date_weekday,
    parsed_date_weeknum,
    count(1) AS "message_amount"
FROM hour_week_cte
GROUP BY
    parsed_date_hour,
    parsed_date_weekday,
    parsed_date_weeknum
ORDER BY
    parsed_date_weeknum,
    parsed_date_weekday

```

Enfim, são diversas as possiblidades para consulta e varia de acordo com o conteúdo e interesse do analista.
